```python
#creating the amino acid triplet codon and single letter code standard codons.
codes = open('gen_code.txt')
genetic_code = codes.readlines()

#genetic_code in dictionary
code ={}
for i in  range(9,len(genetic_code[0])):    
    aa= genetic_code[0][i];  
    codon = genetic_code[2][i] + genetic_code[3][i] + genetic_code[4][i];   
    code[codon] = aa;
```

```python
data = open("rosalind_prot2.txt")
mydata = data.readlines()
pos =0;
for i in mydata:
    aaseq="";
    while (pos < len(i) -2):
        #stroing triplet codon in temporary variable
        tempcodon = i[pos] + i[pos+1] + i[pos+2];
        
        #finding that temp in dictionary keys
        if(tempcodon in code.keys()):
            
            #if found, add its single letter code/value
            aaseq = aaseq + code[tempcodon]
            
        #frameshift by 3 codons
        pos = pos + 3
    print(aaseq)
```
